```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False)#fit regression line,by default fit_reg is true
#x & y lables are auto generated
#scatter plots are used for bi-variate(2) analysis
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='day',data=df1,fit_reg=False)
#hue - sub col to sort the data
#on the basis of unique values of hue col we will get the colors
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df1,fit_reg=False)#how many male and female came
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=df1,fit_reg=False)#how many somkers came
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df1,fit_reg=False,palette='ocean',markers=['o','x'])
#palette-seaborn color attribut, automatically it will asign based on themes,warm,orange,purplc,ocean etc, used in almost all the graphs
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df1,fit_reg=False,row='time')#similarly we have col
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df1,fit_reg=False,row='time',col='smoker')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df1,fit_reg=False,row='day')#who has paid highest tip/bill ---sat 
plt.show()
```
