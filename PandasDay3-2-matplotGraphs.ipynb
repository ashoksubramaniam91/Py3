```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns              #directly it can visualize 
import matplotlib.pyplot as plt    #process and then visualize
```

```python
#seaborn is the extended version of matplotlib
```

```python
%matplotlib inline #for gettinng the graphs inline in the notebook #called as magic function
```

```python
x=[5,4,1,2]
y=[5,6,8,9]
x1=[5,6,8,4]
y1=[6,5,7,8]
```

```python
plt.plot(x,y) #.plot will give line graph
plt.show()#even if it is not given will get the output, but we have to use#also used for aggregating, like for 2 graphs in a single graph
```

```python
plt.plot(x,y)
plt.xlabel('xaxis') #it will b same fr evry graph
plt.ylabel('yaxis')
plt.title('simple line graph')
plt.show()
```

```python
plt.plot(x,y,label='Data') #gives scale for diff lines diff colors
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')
plt.legend() #if label on top is given we should give the use the legend()
plt.show()
```

```python
plt.plot(y,x,label='data1')
plt.plot(x,y,label='Data2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')
plt.legend() 
plt.show()
```

```python
plt.plot(y,x,label='data1')
plt.plot(x1,y1,label='Data2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')
plt.legend() 
plt.show()
```

```python
plt.plot(y,x,label='data1',color='green')#we can give the color using color attribute
plt.plot(x,y,label='Data2',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')
plt.legend() 
plt.show()
```

```python
plt.plot(y,x,label='data1',color='green',linewidth=3)#changing the width of the line
plt.plot(x,y,label='Data2',color='red',linewidth=5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')
plt.legend() 
plt.show()
```

```python
plt.plot(y,x,label='data1',color='green',marker='*')#where ever x-y are meeting 
plt.plot(x,y,label='Data2',color='red',marker='o') #s=square,o=circle
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')
plt.legend() 
plt.show()
```

```python
plt.plot(y,x,label='data1',color='green',marker='*',markersize=10)#marker size 
plt.plot(x,y,label='Data2',color='red',marker='o',markersize=10) 
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')
plt.legend() 
plt.show()
```

```python
x=np.arange(0,10000,100) #sine wave
y=np.sin(x)
```

```python
plt.plot(x,y,label='data1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend() 
plt.show()
```

```python
x2=np.arange(0,10000,100) #co-sine wave (0,10,0.1)
y2=np.cos(x)
```

```python
plt.plot(x2,y2,label='data1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend() 
plt.show()
```

```python
plt.plot(x,y,label='data1',color='green')#sine & cos
plt.plot(x2,y2,label='data2',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend() 
plt.show()
```

```python
Companies=['Infosys','TCS','Wipro','Google','Facebook','GL','Accenture'] #will use this as index
df1=DataFrame([240,200,300,350,280,320,370],index=Companies,columns=['Price'])
plt.bar(df1.index,df1['Price'],label='Data')
plt.legend()
plt.show()
```

```python
#histogram
#it is continouns
#dont need 2 axis we need only x
#it is not necessary that always histograms wont be having gaps, we can have gaps for better visualization
```

```python
BSM=[113,85,90,150,149,88,93,115,135,80,77,82,129]#bloodsugarmales checking sugar levels in males
plt.hist(BSM,bins=4)#bins means 4 parts
plt.show()
```

```python
BSM=[113,85,90,150,149,88,93,115,135,80,77,82,129]#bloodsugarmales checking sugar levels in males
plt.hist(BSM,bins=[75,100,110,115,145,150])#for different width #4 points so 3 bins
plt.show()
```

```python
#for spaces
BSM=[113,85,90,150,149,88,93,115,135,80,77,82,129]#bloodsugarmales checking sugar levels in males
plt.hist(BSM,bins=4,rwidth=0.95) #rwidth will give the space between bins, 0.95 means i want to see 95% of the bin, 1 is default which will have no spaces
plt.show()#0 will show empty
```
