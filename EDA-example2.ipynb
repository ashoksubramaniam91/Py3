```python
#From the box plot we can see that there are outliers and it can be identified using a formula
#q1 - 1.5*(IQR)
#q3 + 1.5*(IQR)
```

```python
#To find outliers
```

```python
for i in df1.columns[0:23]:
    def outliers(i):
        q1 = np.percentile(df1[i],25)
        q3 = np.percentile(df1[i],75)
        iqr = q3-q1
        maxval = q3 + 1.5 * (iqr)
        minval = q1 - 1.5 * (iqr)
        return maxval
        return minval
        
```
